Plot 1)How many emplyees are there in each status catogary

Plot 2) How many active and terminated employees are there in each experience level

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='status',hue='exp_level')
plt.legend(loc='upper left',fontsize=7)
plt.show
```

Observation: Almost active and terminated employees are same but in mid senior and senior have more active employees

Plot 3) Draw a pie chart from the attrition data set the proportion of different termination types among employees

```python
#Pie chart not present in seaborn

#for pie chart first step is  take value and index
val=df['termtype_desc'].value_counts().values
idx=df['termtype_desc'].value_counts().index
print(val)
print(idx)
```

```python
plt.figure(figsize=(6,3))
# plt.pie(x=val)
# plt.pie(x=val,labels=idx)
plt.pie(x=val,labels=idx,autopct='%.2f')
plt.title('Proportion of term_type')
plt.show()
```

Note: We use countplot and piechart for only catogorcal datas if there is is numbers we use KDE (used for continious data only(float))
